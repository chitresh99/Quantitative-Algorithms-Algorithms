Simple Moving Average (SMA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Closing prices
prices = [100, 102, 101, 98, 96, 99, 97, 95, 100, 101]
```

```python
#converting to a panda data frame

data = pd.DataFrame(prices,columns=['Close'])
```

data['SMA_3']: This creates a new column in the DataFrame data named 'SMA_3'.

data['Close']: This accesses the column named 'Close' in the DataFrame data. The 'Close' column typically contains the closing prices of a stock or some other time series data.

.rolling(window=3): This applies a rolling window calculation on the 'Close' column. The window=3 parameter specifies that the window size is 3, meaning the calculation will consider the current and the previous two values (a total of three values).

.mean(): This calculates the mean (average) of the values within each rolling window.

```python
# Calculate the 3-day SMA
data['SMA_3'] = data['Close'].rolling(window=3).mean()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data['Close'], label='Closing Prices')
plt.plot(data['SMA_3'], label='3-Day SMA', linestyle='--')
plt.legend()
plt.title('3-Day Simple Moving Average (SMA)')
plt.xlabel('Day')
plt.ylabel('Price')
plt.show()
```
